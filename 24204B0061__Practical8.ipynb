1. Program using math functions
Power, ceiling, square root, factorial

```python
import math

n = int(input("Enter a number: "))

print("Power (n^2):", math.pow(n, 2))
print("Ceiling:", math.ceil(n))
print("Square Root:", math.sqrt(n))
print("Factorial:", math.factorial(n))
```

3. Program for Half Adder and Full Adder using functions

```python
def half_adder(a, b):
    sum_bit = a ^ b
    carry = a & b
    print("Half Adder -> Sum:", sum_bit, "Carry:", carry)

def full_adder(a, b, cin):
    sum_bit = a ^ b ^ cin
    carry = (a & b) | (b & cin) | (a & cin)
    print("Full Adder -> Sum:", sum_bit, "Carry:", carry)

a = int(input("Enter first bit (0 or 1): "))
b = int(input("Enter second bit (0 or 1): "))

half_adder(a, b)

cin = int(input("Enter carry in (0 or 1): "))
full_adder(a, b, cin)
```
