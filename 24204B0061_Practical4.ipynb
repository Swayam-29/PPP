Develop python program to perform following operations on List:
a) Create
b) Access
c) Update
d) Delete elements from list

```python
## marks = []

while True:
    print("\n--- Student Management System ---")
    print("1. Enter student marks")
    print("2. Display all marks")
    print("3. Total and Average marks")
    print("4. Highest and Lowest marks")
    print("5. Passed and Failed students")
    print("6. Distinction students (>=75)")
    print("7. Remove duplicate marks")
    print("8. Replace failed marks with grace (40)")
    print("9. Search a mark")
    print("10. Sort marks & reverse")
    print("11. Remove a student record")
    print("12. Find position of a mark")
    print("13. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        n = int(input("How many students? "))
        for i in range(n):
            m = int(input("Enter mark: "))
            marks.append(m)

    elif choice == 2:
        print("Student Marks:", marks)

    elif choice == 3:
        if len(marks) > 0:
            total = sum(marks)
            avg = total / len(marks)
            print("Total =", total)
            print("Average =", avg)
        else:
            print("No marks available")

    elif choice == 4:
        if len(marks) > 0:
            print("Highest =", max(marks))
            print("Lowest =", min(marks))
        else:
            print("No marks available")

    elif choice == 5:
        passed = []
        failed = []
        for m in marks:
            if m >= 40:
                passed.append(m)
            else:
                failed.append(m)
        print("Passed students:", passed)
        print("Failed students:", failed)

    elif choice == 6:
        distinction = []
        for m in marks:
            if m >= 75:
                distinction.append(m)
        print("Distinction students:", distinction)

    elif choice == 7:
        marks = list(set(marks))
        print("Duplicates removed:", marks)

    elif choice == 8:
        for i in range(len(marks)):
            if marks[i] < 40:
                marks[i] = 40
        print("Grace marks applied:", marks)

    elif choice == 9:
        s = int(input("Enter mark to search: "))
        if s in marks:
            print("Mark found")
        else:
            print("Mark not found")

    elif choice == 10:
        marks.sort()
        print("Sorted marks:", marks)
        print("Reverse order:", marks[::-1])

    elif choice == 11:
        r = int(input("Enter mark to remove: "))
        if r in marks:
            marks.remove(r)
            print("Record removed")
        else:
            print("Mark not found")

    elif choice == 12:
        p = int(input("Enter mark: "))
        if p in marks:
            print("Position (index):", marks.index(p))
        else:
            print("Mark not found")

    elif choice == 13:
        print("Exiting program...")
        break

    else:
        print("Invalid choice")
```
