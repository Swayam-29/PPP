1. Write a Python NumPy program to create a 3x3 identity matrix.

```python
import numpy as np

identity_matrix = np.eye(3)

print(identity_matrix)
```

2. Write a Python NumPy program to create a 4x4 matrix with values ranging from 2 to 10

```python
import numpy as np

matrix = np.arange(2, 18).reshape(4, 4)

print(matrix)
```

3. Program to create a user-defined package
Step 1 — Folder Structure
Step 2 — ops.py
Step 3 — Main Program

```python
import os
os.makedirs("mypackage", exist_ok=True)
```

```python
open("mypackage/__init__.py", "w").close()
```

```python
%%writefile mypackage/ops.py

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b
```

```python
from mypackage import ops

print("Addition:", ops.add(5, 3))
print("Multiplication:", ops.multiply(5, 3))
```
