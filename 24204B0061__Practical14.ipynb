1. Program for Stack Operations Push, Pop, Display 

```python
stack = []

while True:
    print("\n1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        item = input("Enter element to push: ")
        stack.append(item)
        print("Element pushed")

    elif choice == 2:
        if len(stack) == 0:
            print("Stack is empty")
        else:
            print("Popped element:", stack.pop())

    elif choice == 3:
        print("Stack:", stack)

    elif choice == 4:
        break

    else:
        print("Invalid choice")
```

 2. Program for Queue Operations Insert, update, delete, display

```python
queue = []

while True:
    print("\n1. Insert")
    print("2. Update")
    print("3. Delete")
    print("4. Display")
    print("5. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        item = input("Enter element to insert: ")
        queue.append(item)
        print("Element inserted")

    elif choice == 2:
        pos = int(input("Enter position to update: "))
        if 0 <= pos < len(queue):
            queue[pos] = input("Enter new value: ")
            print("Element updated")
        else:
            print("Invalid position")

    elif choice == 3:
        if len(queue) == 0:
            print("Queue is empty")
        else:
            print("Deleted element:", queue.pop(0))

    elif choice == 4:
        print("Queue:", queue)

    elif choice == 5:
        break

    else:
        print("Invalid choice")
```

3. Program to implement both Stack & Queue in same code

```python
stack = []
queue = []

while True:
    print("\n1. Stack Push")
    print("2. Stack Pop")
    print("3. Stack Display")
    print("4. Queue Insert")
    print("5. Queue Delete")
    print("6. Queue Display")
    print("7. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        item = input("Enter element to push into stack: ")
        stack.append(item)
        print("Pushed into stack")

    elif choice == 2:
        if len(stack) == 0:
            print("Stack is empty")
        else:
            print("Popped from stack:", stack.pop())

    elif choice == 3:
        print("Stack:", stack)

    elif choice == 4:
        item = input("Enter element to insert into queue: ")
        queue.append(item)
        print("Inserted into queue")

    elif choice == 5:
        if len(queue) == 0:
            print("Queue is empty")
        else:
            print("Deleted from queue:", queue.pop(0))

    elif choice == 6:
        print("Queue:", queue)

    elif choice == 7:
        break

    else:
        print("Invalid choice")
```
