1. Program for Linear Search

```python
data = list(map(int, input("Enter elements separated by space: ").split()))
key = int(input("Enter element to search: "))

found = False

for i in range(len(data)):
    if data[i] == key:
        print("Element found at position", i + 1)
        found = True
        break

if not found:
    print("Element not found")
```

2. Program for Binary Search

```python
data = list(map(int, input("Enter sorted elements separated by space: ").split()))
key = int(input("Enter element to search: "))

low = 0
high = len(data) - 1
found = False

while low <= high:
    mid = (low + high) // 2
    
    if data[mid] == key:
        print("Element found at position", mid + 1)
        found = True
        break
    elif data[mid] < key:
        low = mid + 1
    else:
        high = mid - 1

if not found:
    print("Element not found")
```

3. Program to search for a character in a string (Linear Search logic)

```python
text = input("Enter a string: ")
ch = input("Enter character to search: ")

found = False

for i in range(len(text)):
    if text[i] == ch:
        print("Character found at position", i + 1)
        found = True
        break

if not found:
    print("Character not found")
```
