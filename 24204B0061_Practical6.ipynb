Shopping Cart System – Set Based Programming Problem
Design and implement a Shopping Cart System using a set data structure to manage product IDs added to a cart.
The system should provide the following functionalities:
•	Allow the user to add product IDs to the shopping cart using a set.
•	Display all products currently present in the cart.
•	Automatically ensure that duplicate product IDs are not allowed in the cart.
•	Display the total number of unique products in the cart.
•	Identify and display the highest and lowest product ID in the cart.
•	Provide a search feature to check whether a specific product ID exists in the cart.
•	Remove a product ID from the cart when an item is deleted.
•	Display all premium products whose IDs are greater than or equal to a given value.
•	Perform union and intersection operations between carts of two different users.
•	Display the shopping cart items in sorted order.
Use appropriate set operations and functions to implement the above features.



```python
n1 = int(input("Enter number of products for Cart 1: "))

cart1_items = []
for i in range(n1):
    pid = int(input())
    cart1_items.append(pid)

cart1 = set(cart1_items)

premium_limit = int(input("Enter premium product ID limit: "))
search_id = int(input("Enter product ID to search: "))
remove_id = int(input("Enter product ID to remove: "))

n2 = int(input("Enter number of products for Cart 2: "))

cart2_items = []
for i in range(n2):
    pid = int(input())
    cart2_items.append(pid)

cart2 = set(cart2_items)

print(cart1)
print(len(cart1))

if cart1:
    print(max(cart1))
    print(min(cart1))

if search_id in cart1:
    print("Found")
else:
    print("Not Found")

cart1.discard(remove_id)
print(cart1)

premium_products = set()
for pid in cart1:
    if pid >= premium_limit:
        premium_products.add(pid)
print(premium_products)

print(cart1 | cart2)
print(cart1 & cart2)

print(sorted(cart1))
```
