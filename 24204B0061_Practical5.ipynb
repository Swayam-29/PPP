Library Inventory System – Tuple Based Programming Problem
Design and implement a Library Inventory System that uses a tuple data structure to store and manage information about book IDs (or book prices / accession numbers).
The system should provide the following functionalities:
•	Allow the librarian to enter book data and store it in a tuple.
•	Display all book records stored in the system.
•	Calculate and display the total number of books and the average value (price or ID-based calculation).
•	Identify and display the highest and lowest book values in the collection.
•	Display a list of premium books whose value is greater than or equal to a given limit.
•	Provide a search feature to check whether a particular book value exists in the tuple.
•	Count how many times a specific book value appears in the tuple.
•	Convert the tuple into a list to remove duplicate book records, then convert it back into a tuple.
•	Display the tuple in reverse order.
•	Find and display the position (index) of a particular book value in the tuple.
Use appropriate tuple operations and built-in functions to implement the above features, keeping in mind that tuples are immutable.

```python
n = int(input("Enter number of books: "))

book_list = []
for i in range(n):
    value = int(input("Enter book value (ID/Price): "))
    book_list.append(value)

books = tuple(book_list)

search_value = int(input("Enter book value to search: "))
premium_limit = int(input("Enter premium book limit: "))

print("\n--- Library Inventory Report ---")

# Display all book records
print("Book Records:", books)

# Total number of books and average value
total_books = len(books)
average_value = sum(books) / total_books
print("Total Books:", total_books)
print("Average Value:", average_value)

# Highest and lowest value
print("Highest Book Value:", max(books))
print("Lowest Book Value:", min(books))

# Premium books
premium_books = []
for b in books:
    if b >= premium_limit:
        premium_books.append(b)
print("Premium Books:", premium_books)

# Search for a book value
if search_value in books:
    print("Search Result:", search_value, "found in library")
else:
    print("Search Result:", search_value, "not found in library")

# Count occurrences
count = books.count(search_value)
print("Occurrence Count of", search_value, ":", count)

# Remove duplicates (tuple → list → tuple)
unique_books = tuple(set(books))
print("Books after removing duplicates:", unique_books)

# Reverse tuple
print("Books in Reverse Order:", books[::-1])

# Find position (index)
if search_value in books:
    print("Position (Index) of", search_value, ":", books.index(search_value))
else:
    print("Position: Not available")
```
