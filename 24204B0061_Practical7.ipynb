Bank Account System – Dictionary Based Programming Problem
Design and implement a Bank Account Management System using a dictionary data structure, where the account number is the key and the account balance is the value.
The system should provide the following functionalities:
•	Allow the admin to create bank accounts by storing account numbers and balances in a dictionary.
•	Display all account details stored in the system.
•	Calculate and display the total and average balance of all accounts.
•	Identify and display the highest and lowest account balances.
•	Display accounts with high-value balances above a specified limit.
•	Provide a search feature to check whether a particular account number exists.
•	Update the account balance after deposit or withdrawal.
•	Remove an account record when an account is closed.
•	Display all account numbers and balances separately.
•	Sort and display accounts based on account numbers.
Use appropriate dictionary methods and operations to implement the above features.

```python
accounts = {}

n = int(input("Enter number of accounts: "))
for i in range(n):
    acc_no = int(input("Enter account number: "))
    bal = float(input("Enter balance: "))
    accounts[acc_no] = bal

limit = float(input("Enter high balance limit: "))
search_acc = int(input("Enter account number to search: "))
update_acc = int(input("Enter account number to update: "))
amount = float(input("Enter amount (+deposit / -withdraw): "))
remove_acc = int(input("Enter account number to remove: "))

print("\nAll Accounts:")
print(accounts)

total_balance = sum(accounts.values())
average_balance = total_balance / len(accounts) if accounts else 0
print("Total Balance:", total_balance)
print("Average Balance:", average_balance)

if accounts:
    print("Highest Balance:", max(accounts.values()))
    print("Lowest Balance:", min(accounts.values()))

high_value_accounts = {}
for k, v in accounts.items():
    if v > limit:
        high_value_accounts[k] = v
print("High Value Accounts:", high_value_accounts)

if search_acc in accounts:
    print("Account Found:", search_acc)
else:
    print("Account Not Found:", search_acc)

if update_acc in accounts:
    accounts[update_acc] += amount
print("Accounts after update:", accounts)

accounts.pop(remove_acc, None)
print("Accounts after removal:", accounts)

print("Account Numbers:")
print(list(accounts.keys()))

print("Balances:")
print(list(accounts.values()))

print("Sorted Accounts by Account Number:")
print(dict(sorted(accounts.items())))
```
